# Importing libraries that we will use

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## reading data

```python
data=pd.read_csv('http://bit.ly/w-data')
data.head()
```

# showing data in a 2D graph

```python
data.plot(x='Hours',y='Scores',style='o',color='black')
plt.title('Hours vs Scores')
plt.xlabel('Hours')
plt.ylabel('Score')
plt.show
```

# splitting data into features and labels 

```python
x=data.iloc[:,0].values.reshape(-1,1)
y=data.iloc[:,-1].values.reshape(-1,1)
```

# Building linear regression model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
print('Fitting Complete')
```

# showing best fitting line

```python
bfl=model.coef_*x+model.intercept_
plt.scatter(x,y)
plt.plot(x,bfl,color='black')
plt.show
```

# model testing

```python
h=float(input('Hours Per Day: '))
hrs=np.array([[h]])
score_pred=model.predict(hrs)
print(f'Predicted Score Is {score_pred[0][0].round(2)}')
```

# model evaluating

y_pred = model.predict(x_test)

```python
from sklearn import metrics 
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
```
